```python
import pandas as pd
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# 한글 폰트 설정
plt.rcParams['font.family'] = 'AppleGothic'

# CSV 파일 불러오기
ssg = pd.read_csv('./ssg_0.4.csv')

# 날짜를 datetime 형식으로 변환
ssg['date'] = pd.to_datetime(ssg['date'])

# 2021년 5월에 해당하는 데이터 필터링
may_2021_data = ssg[(ssg['date'].dt.year == 2021) & (ssg['date'].dt.month == 5)]

# 각 row의 주요 단어 리스트 추출
all_words = []
for indices in may_2021_data['indices']:
    # 문자열 형태의 리스트를 실제 리스트로 변환 (예: "[word1, word2]" -> ["word1", "word2"])
    words = indices.strip('[]').replace("'", "").split(', ')
    all_words.extend(words)

# 단어 빈도 계산
word_freq = Counter(all_words)

# 빈도가 높은 상위 단어 제외
exclude_words = ['제외할단어1', '제외할단어2', '제외할단어3']  # 예시: 제외할 단어들을 직접 추가하세요
for word in exclude_words:
    del word_freq[word]

# 워드클라우드 생성
wordcloud = WordCloud(font_path='/Library/Fonts/AppleGothic.ttf', width=800, height=400, background_color='white', colormap='Oranges').generate_from_frequencies(word_freq)

# 워드클라우드 시각화
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('ssg 2021년 5월 키워드 워드클라우드')
plt.show()
```

```python
import pandas as pd
from itertools import combinations
from collections import defaultdict
from gensim.models import Word2Vec
```

```python
# 단어 쌍 카운트 딕셔너리 생성
word_pairs = defaultdict(int)

# 각 행의 단어 쌍 추출 및 카운트
for row in df['indices']:
    words = row.split(', ')
    for pair in combinations(words, 2):
        word_pairs[pair] += 1

# Word2Vec 모델 학습을 위한 데이터 전처리
sentences = [row.split(', ') for row in df['indices']]

# Word2Vec 모델 학습
model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, sg=1)

# 모델 사용 예시: 단어 벡터 얻기
word_vector = model.wv['김하성']

# 모델 사용 예시: 단어 간 유사도 계산
similarity = model.wv.similarity('김하성', '크로넨워스')

print(f"김하성 벡터: {word_vector}")
print(f"김하성-크로넨워스 유사도: {similarity}")
```
