```python
import pandas as pd

# CSV 파일 불러오기
df_embedding = pd.read_csv('./embedding.csv')
df_date = pd.read_csv('./date.csv')
```

```python
merged_df = pd.merge(df_embedding, df_date, left_on='Name', right_on='content', how='inner')
```

```python
# indices_x 컬럼 삭제
merged_df.drop('indices_x', axis=1, inplace=True)

# indices_y 컬럼의 이름을 indices로 변경
merged_df = merged_df.rename(columns={'indices_y': 'indices'})
```

```python
merged_df
```

```python
park_donghee_entries = merged_df[merged_df['date'].str.contains('기자명')]
merged_df.drop(park_donghee_entries.index, inplace=True)
```

```python
merged_df['date'] = pd.to_datetime(merged_df['date'].str.replace('입력 ', ''), format='%Y.%m.%d %H:%M')
```

```python
merged_df
```

```python
# 병합된 데이터프레임을 새로운 CSV 파일로 저장
merged_df.to_csv('final.csv', index=False)
```
