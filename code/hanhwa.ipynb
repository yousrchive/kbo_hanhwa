```python
import pandas as pd
import matplotlib.pyplot as plt

# 날짜를 datetime 형식으로 변환
hanhwa['date'] = pd.to_datetime(hanhwa['date'])

# 연도와 month 추출
hanhwa['year'] = hanhwa['date'].dt.year
hanhwa['month'] = hanhwa['date'].dt.month

# year와 month, sentiment를 기준으로 groupby하고 count
grouped = hanhwa.groupby(['year', 'month', 'sentiment']).size().unstack(fill_value=0)

# 시각화
fig, ax = plt.subplots(figsize=(16, 8))  # 가로로 길게 설정

# 누적막대그래프
grouped.plot(kind='bar', stacked=True, ax=ax, color=['blue', 'red'])

# 그래프 설정
ax.set_ylabel('Number of News Articles')
ax.set_xlabel('Year-Month')
ax.set_title('Monthly Sentiment Distribution by News Count')
ax.legend(title='Sentiment')

# x축 레이블 설정
labels = [f'{year}-{month:02d}' for year, month in grouped.index]
ax.set_xticklabels(labels, rotation=45, ha='right')

# 각 막대 위에 전체 뉴스 개수 표시
for i in range(len(grouped)):
    total = grouped.iloc[i].sum()
    ax.text(i, total + 0.1, str(total), ha='center', va='bottom')

# 레이아웃 조정
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 함수 정의
def plot_sentiment_distribution(file_path, sentiment_colors):
    # CSV 파일 불러오기
    df = pd.read_csv(file_path)

    # 날짜를 datetime 형식으로 변환
    df['date'] = pd.to_datetime(df['date'])

    # 연도와 month 추출
    df['year'] = df['date'].dt.year
    df['month'] = df['date'].dt.month

    # year와 month, sentiment를 기준으로 groupby하고 count
    grouped = df.groupby(['year', 'month', 'sentiment']).size().unstack(fill_value=0)

    # 시각화
    fig, ax = plt.subplots(figsize=(16, 8))  # 가로로 길게 설정

    # 누적막대그래프
    grouped.plot(kind='bar', stacked=True, ax=ax, color=sentiment_colors)

    # 그래프 설정
    ax.set_ylabel('Number of News Articles')
    ax.set_xlabel('Year-Month')
    ax.set_title(f'Monthly Sentiment Distribution by News Count - {file_path}')
    ax.legend(title='Sentiment')

    # x축 레이블 설정
    labels = [f'{year}-{month:02d}' for year, month in grouped.index]
    ax.set_xticklabels(labels, rotation=45, ha='right')

    # 각 막대 위에 전체 뉴스 개수 표시
    for i in range(len(grouped)):
        total = grouped.iloc[i].sum()
        ax.text(i, total + 0.1, str(total), ha='center', va='bottom')

    # 레이아웃 조정
    plt.tight_layout()
    plt.show()

# 파일 경로와 색상 설정
file_paths = ['./hanhwa_0.4.csv', './doosan_0.4.csv', './lg_0.35.csv', './nc_0.4.csv', './ssg_0.4.csv']
sentiment_colors = ['blue', 'red']  # 감성에 따른 색상 설정 (긍정: blue, 부정: red)

# 각 파일에 대해 그래프 그리기
for file_path in file_paths:
    plot_sentiment_distribution(file_path, sentiment_colors)
```
